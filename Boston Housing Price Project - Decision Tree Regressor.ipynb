```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
boston = pd.read_csv('Boston1.csv')
```

```python
boston.head()
```

```python
plt.scatter(x=boston['rm'], y=boston['medv'], color = 'brown')
plt.xlabel("Average Number of Rooms per dwelling")
plt.ylabel("Median Value of Home")
plt.show()
```

```python
#Extracting features and targets from the dataframe

x = pd.DataFrame(boston['rm'])
y = pd.DataFrame(boston['medv'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred, y_test)
rmse = np.sqrt(mse)
print(mse)
print(rmse)
```

```python
#Adding more features to the model

x = pd.DataFrame(boston[['rm', 'lstat', 'age']])
y = pd.DataFrame(boston['medv'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
mse = mean_squared_error(y_pred, y_test)
rmse = np.sqrt(mse)
print(mse)
print(rmse)
```

Since the second model has a lower RMSE value, we can predict that the second model is better than the first model.
